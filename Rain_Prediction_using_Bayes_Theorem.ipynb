Aim: Evaluate Bayes Rules for Rain predication from given scenario.

```python
# Given probabilities
p_cloudy = 0.40
p_rain = 0.20
p_cloudy_given_rain = 0.85

# Apply Bayes Theorem
# P(Rain | Cloudy) = (P(Rain) * P(Cloudy | Rain)) / P(Cloudy)
p_rain_given_cloudy = (p_rain * p_cloudy_given_rain) / p_cloudy

# Output
print("Bayes Theorem - Rain Prediction")
print("--------------------------------")
print("Given:")
print("P(Cloudy) =", p_cloudy)
print("P(Rain) =", p_rain)
print("P(Cloudy | Rain) =", p_cloudy_given_rain)

print("\nUsing Bayes Theorem:")
print("P(Rain | Cloudy) = (P(Rain) * P(Cloudy | Rain)) / P(Cloudy)")

print("\nCalculated Probability:")
print("P(Rain | Cloudy) =", round(p_rain_given_cloudy, 3))

print("\nIn Percentage:")
print("P(Rain | Cloudy) =", round(p_rain_given_cloudy * 100, 2), "%")
```
