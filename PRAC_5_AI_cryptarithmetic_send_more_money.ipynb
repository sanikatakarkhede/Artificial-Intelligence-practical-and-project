Aim: Write a code to Implement Cryptarithmetic puzzle (SEND +
MORE=MONEY) using the concept of constraint satisfaction concept.

```python
from itertools import permutations

letters = ('S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y')  # 0 -> O fix

digits = range(10)

for perm in permutations(digits, len(letters)):

    S, E, N, D, M, O, R, Y = perm   # 0 -> O fix

    if S == 0 or M == 0:
        continue

    SEND = 1000*S + 100*E + 10*N + D
    MORE = 1000*M + 100*O + 10*R + E
    MONEY = 10000*M + 1000*O + 100*N + 10*E + Y

    if SEND + MORE == MONEY:
        print("Solution Found!")
        print(f"S={S}, E={E}, N={N}, D={D}, M={M}, O={O}, R={R}, Y={Y}")
        print(f"{SEND} + {MORE} = {MONEY}")
        break
```
